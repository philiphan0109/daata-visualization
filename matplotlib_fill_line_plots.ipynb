```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("data/line_plot_fill_data.csv")

ages = data["Age"]
dev_salaries = data["All_Devs"]
js_salaries = data["JavaScript"]
py_salaries = data["Python"]
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 60000
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate=True, alpha = 0.5, label = "Above All Devs")

plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries < dev_salaries), interpolate=True, alpha = 0.5, label = "Below All Devs")

plt.legend()
plt.tight_layout()

plt.title("Median Salaries by Age")
plt.xlabel("Age")
plt.ylabel("Median Salary")
plt.show()
```
