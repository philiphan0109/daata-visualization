```python
from matplotlib import pyplot as plt
import random
```

```python
x = [i for i in range (15, 30)]
y = [random.randint(30000, 50000) for i in x]

plt.plot(x, y)

x_python = [i for i in range (15, 30)]
y_python = [random.randint(30000, 40000) for i in x]

plt.plot(x_python, y_python)

plt.title("Average income by age for python developers")
plt.xlabel("Age")
plt.ylabel("Income ($)")
plt.show()
```
